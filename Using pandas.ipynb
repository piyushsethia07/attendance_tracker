```python
import pandas as pd
from openpyxl import load_workbook

# Specify the Excel file path
excel_file_path = 'Mel Sat.01AdrianWhite August 2023.xlsx'

# Load the Excel workbook
wb = load_workbook(excel_file_path, data_only=True)  # Use data_only=True to read evaluated values

# Create an empty list to store the data
data = []

# Iterate through all sheets in the workbook
for sheet in wb.sheetnames:
    # Read each sheet as a DataFrame
    df = pd.read_excel(excel_file_path, sheet_name=sheet, header=None)
    df = df.dropna(how='all')
    data.append(df)

# Now, 'data' contains a list of DataFrames, each representing a sheet in the Excel file
# Each DataFrame contains all the values and formulas present in the corresponding sheet
```
